```python
import pandas as pd
```

```python
df = pd.read_csv("Artists.csv")
```

### Temel istatistik bilgileri görüntüleme

```python
summary_stats = df.describe()
print("Temel İstatistiksel Bilgilerin Görüntülenmesi:")
print(summary_stats)
```

### Belirli bir sütunun içerisinde  istatistiksel bilgilerini görüntüleme

```python
column_stats = df['Name'].describe()
print("\nSadece Belirli Bir Sütunun İstatistiksel Bilgileri:")
print(column_stats)
```

# Belirli bir sütundaki eşsiz değerleri ve sayılarını görüntüleme

```python
unique_values = df['Followers'].value_counts()
print("\nBenzersiz Değerler ve Sayıları:")
print(unique_values)
```

# Korelasyon matrisini görüntüleme

```python
correlation_matrix = df.corr()
print("\nKorelasyon Matrisi:")
print(correlation_matrix)
```

# Belirli bir sütunun histogramını çiz

```python
df['Country'].hist()
```

```python
import matplotlib.pyplot as plt
```

```python
df['Country'].hist()
plt.title('Country Histogramı')
plt.xlabel('Değerler')
plt.ylabel('Frekans')
plt.show()
```

# Belirli bir sütunun kutu grafiğini çizmek

```python
df.boxplot(column='Age')
plt.title('Age Kutu Grafiği')
plt.ylabel('Değerler')
plt.show()
```
